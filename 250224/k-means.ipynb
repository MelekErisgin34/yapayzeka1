```python
# Gerekli kütüphaneleri yükleyelim
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

# 1. Adım: Veri setini yükle
iris = load_iris()
X = iris.data  # Özellikler (features)
y = iris.target  # Gerçek etiketler (gerçek türler)

# 2. Adım: K-means algoritmasını uygulayalım
kmeans = KMeans(n_clusters=3, random_state=42)  # 3 küme oluşturuyoruz, çünkü İris veri setinde 3 tür var.
kmeans.fit(X)

# Kümelerin etiketlerini alalım
labels = kmeans.labels_

# 3. Adım: Sonuçları görselleştirelim
# PCA kullanarak 4 boyutlu veriyi 2 boyuta indireceğiz, çünkü görselleştirme için 2D gereklidir
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

plt.figure(figsize=(8,6))

# Gerçek etiketler ile görselleştirme (gerçek türler)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', label='Gerçek Etiketler', alpha=0.5)
plt.title("Gerçek Etiketlerle Görselleştirme")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.colorbar()
plt.show()

# K-means etiketleri ile görselleştirme (tahmin edilen kümeler)
plt.figure(figsize=(8,6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='viridis', label='K-means Etiketleri', alpha=0.5)
plt.title("K-means ile Görselleştirme")
plt.legend()
plt.show()

print(f"Kümelenmiş Etiketler: {kmeans.labels_}")
```
