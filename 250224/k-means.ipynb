```python
# Gerekli ktphaneleri ykleyelim
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

# 1. Adm: Veri setini ykle
iris = load_iris()
X = iris.data  # zellikler (features)
y = iris.target  # Gerek etiketler (gerek trler)

# 2. Adm: K-means algoritmasn uygulayalm
kmeans = KMeans(n_clusters=3, random_state=42)  # 3 kme oluturuyoruz, nk ris veri setinde 3 tr var.
kmeans.fit(X)

# Kmelerin etiketlerini alalm
labels = kmeans.labels_

# 3. Adm: Sonular grselletirelim
# PCA kullanarak 4 boyutlu veriyi 2 boyuta indireceiz, nk grselletirme iin 2D gereklidir
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

plt.figure(figsize=(8,6))

# Gerek etiketler ile grselletirme (gerek trler)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', label='Gerek Etiketler', alpha=0.5)
plt.title("Gerek Etiketlerle Grselletirme")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.colorbar()
plt.show()

# K-means etiketleri ile grselletirme (tahmin edilen kmeler)
plt.figure(figsize=(8,6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='viridis', label='K-means Etiketleri', alpha=0.5)
plt.title("K-means ile Grselletirme")
plt.legend()
plt.show()

print(f"Kmelenmi Etiketler: {kmeans.labels_}")
```
